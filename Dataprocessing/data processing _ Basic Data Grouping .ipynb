Basic data grouping

```python
import pandas as pd
import numpy as np
```

```python
#load the data into a DataFrame
data = pd.read_csv('endangeredLang.csv')
```

```python
data.head(6)
```

```python
large = data.sort_values(by = 'Number of speakers', ascending = False)
```

```python
large.head(6)
```

```python
# sort in descending order 
small = data.sort_values(by= 'Number of speakers', ascending=True)
```

```python
small.head(6)
```

```python
byStatus  = data.groupby('Degree of endangerment')
```

```python
byStatus.head(n=6)
```

```python
byStatus.count()
```

```python
# grouping data by qualitative cloumn 
data.groupby(['Degree of endangerment']).count()
```

```python
# grouping data basis on two qualitative columns
data.groupby(['Countries','Degree of endangerment'])
```

```python
data.groupby(['Countries','Degree of endangerment']).count()
```

```python
# group no of speakers by group labels(row names) from countries 
x = data['Number of speakers'].groupby(data['Countries'])
x.head(n =10)
```

 Lambda Functions

Identifies which of the languages have fewer than 5k speakers

```python
data['vfew'] = data['Number of speakers'].apply(lambda x: x <= 5000)
# vfew column has created and added to the dataframe at the end  by true or false values 
```

```python
data.head(6)
```

```python
data[data['Number of speakers'] <=5000]
```

```python
groupby_5k = data.groupby(['vfew'])
```

```python
groupby_5k.size()
# fewer than 5k is 1843 , more than 5k is 879 
```

```python
groupby_5kc = data.groupby(['vfew','Countries'])
```

```python
groupby_5kc.size()
```

```python
data['Degree of endangerment'].value_counts()
```

Indexing and Subsetting 

```python
data[2:] # select all entries from index 2 onwards
```

```python
data[::2]  # select entries data at even index locations 
```

```python
data.iloc[0:8,0:7] # specifies the ranges of rows and columns , iloc[row slicing, column slicing]
```

```python
# select all the columns for rows of index values 0 and 10
data.loc[[0,10], :]
```

```python
# isolate rows and columns
data[['Countries','Name in English']][4:8]
```

```python
# subset data on the basis of row labels
data.head(4)
```

```python
data[data.Countries == 'Italy'].head(4)
```

```python
data[data.Countries == "India"]
```
